```python
menu ={"qahva": 12000,"choy": 8000,"sok": 10000}
print("Menyuga xush kelibsiz!")
for item, price in menu.items():
    print(f"{item.title()}: {price} so'm")
buyurtma = input("Nima buyurtma qilmoqchisiz? ").lower()
if buyurtma in menu:
    try:
        soni = int(input("Nechta stakan olasiz? "))

        if soni <= 0:
            print("Xato! Musbat son kiriting.")
        else:
            narx = menu[buyurtma] * soni
            if soni > 3:
                narx = narx * 0.9
                print("Sizga 10% chegirma berildi!")
            print("\n--- Buyurtma tafsilotlari ---")
            print(f"Mahsulot: {buyurtma.title()}")
            print(f"Soni: {soni}")
            print(f"Umumiy narx: {int(narx)} so'm")

    except ValueError:
        print("Xato! Iltimos, son kiriting.")
else:
    print("Kechirasiz, bunday mahsulot mavjud emas.")
```

```python
import random
amal = input("Amalni tanlang (+, -, *, //): ")
if amal not in ["+", "-", "*", "//"]:
    print(" Noto‘g‘ri amal kiritdingiz!")
    exit()
son1 = random.randint(1, 10)
son2 = random.randint(1, 10)
savol = f"{son1} {amal} {son2} = ? "
javob = input(savol)
try:
    javob = int(javob)
except ValueError:
    print("Iltimos, butun son kiriting!")
    exit()
if amal == "+":
    togri = son1 + son2
elif amal == "-":
    togri = son1 - son2
elif amal == "*":
    togri = son1 * son2
elif amal == "//":
    togri = son1 // son2
if javob == togri:
    print("To‘g‘ri!")
else:
    print(f"Xato! To‘g‘ri javob: {togri}")
```

```python
balans = 1000
pin = "1234"
print("=== Bankomat Dasturi ===")
print("1. Balansni tekshirish")
print("2. Pul qo‘yish (deposit)")
print("3. Pul yechish (withdraw)")
print("4. PIN kodni o‘zgartirish")
print("5. Chiqish")
amal = input("Amalni tanlang (1-5): ")
if amal == "1":
    print(f"Sizning balansingiz: {balans} so'm")
elif amal == "2":
    try:
        summa = int(input("Qancha pul qo'yasiz? "))
        if summa > 0:
            balans += summa
            print(f"Pul qo'yildi. Yangi balans: {balans} so'm")
        else:
            print("Iltimos, musbat son kiriting!")
    except ValueError:
        print("Iltimos, butun son kiriting!")
elif amal == "3":
    try:
        summa = int(input("Qancha pul yechmoqchisiz? "))
        if summa > balans:
            print("Balans yetarli emas!")
        elif summa <= 0:
            print("Iltimos, musbat son kiriting!")
        else:
            balans -= summa
            print(f"Pul yechildi. Yangi balans: {balans} so'm")
    except ValueError:
        print("Iltimos, butun son kiriting!")
elif amal == "4":
    eski_pin = input("Joriy PIN kodni kiriting: ")
    if eski_pin == pin:
        yangi_pin = input("Yangi PIN kodni kiriting: ")
        tasdiq = input("Yangi PIN kodni qaytadan kiriting: ")
        if yangi_pin == tasdiq:
            pin = yangi_pin
            print("PIN kodi muvaffaqiyatli o'zgartirildi!")
        else:
            print("PIN kod mos kelmadi!")
    else:
        print("Noto‘g‘ri PIN kodi!")
elif amal == "5":
    print("Chiqish...")
else:
    print("Noto‘g‘ri tanlov!")
```
