```python
def jang(player_health=None,enemy_attack=None,armor=None):
    if player_health==None or enemy_attack==None or armor==None:
        return 'qiymatlar toliq kiritilmadi'
    elif isinstance(player_health,bool) or isinstance(enemy_attack,bool) or isinstance(armor,bool):
        return 'faqat son kiriting'
    try:
        player_health=int(player_health)
        enemy_attack=int(enemy_attack)
        armor=int(armor)
    except:
        return 'kiritlgan qiymatlar faqat son bolishi kerak'
    else:
        yangi_health=player_health - (enemy_attack - armor)
    if yangi_health<=0:
        return 'O‘yin tugadi! Siz mag‘lub bo‘ldingiz'
    elif yangi_health<20:
        return "Og‘ir jarohat! Shifobaxsh ichimlik izlang."
    else:
        return " Siz jangni davom ettiryapsiz!"
print(jang(True,34,'897'))
```

```python
def viktorina(amal=None):
    if  amal is None:
        return 'amalni kiritmadingiz(+,-,//,*)'
    elif amal not in ('+','-','*','//'):
        return 'notogri amal'
    import random
    son1=random.randint(1,10)
    son2=random.randint(1,10)
    try:
        javob=int(input(f'{son1}{amal}{son2}= '))
    except:
        return 'javob faqat son bolishi kerak'
    if amal=='+':
        togri=son1+son2
    elif amal=='-':
        togri=son1-son2
    elif amal=='*':
        togri=son1*son2
    elif amal=='//':
        togri=son1//son2
    if javob==togri:
        return 'tog\'ri'
    else:
        return f'Xato! tog\'ri javob {togri}'
print(viktorina('*'))
```
