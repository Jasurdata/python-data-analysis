```python
jumla='boy ota kitobi muallifi Robert Kiyosaki. Robert Kiyosaki boy ota kitobini yozgan'
def word_frequency(file_name):
    with open(file_name, "r") as f:
        text = f.read()
    words = text.split()
    freq = []
    for w in words:
        exists = False
        for i in freq:
            if i[0] == w:
                i[1] += 1
                exists = True
        if not exists:
            freq.append([w, 1])
    return freq
file_name = input("Fayl nomini kiriting: ")
result = word_frequency(file_name)
for w in result:
    print(w[0], "-", w[1])
word_frequency(jumla)
```

```python
def char_histogram(filename):
    # Faylni o'qish
    with open(filename, "r") as f:
        text = f.read()

    # Harflarni sanash
    hist = []  # [harf, soni] saqlanadi
    for ch in text:
        found = False
        for item in hist:
            if item[0] == ch:
                item[1] += 1
                found = True
        if not found:
            hist.append([ch, 1])

    return hist

# --- Ishlatish ---
file_name = input("Fayl nomini kiriting: ")
result = char_histogram(file_name)

# Natijani chiqarish
for item in result:
    print(item[0], "-", item[1])
char_histogram(jumla)
```

```python
import random

def survey_analysis(filename):
    ages = []
    with open(filename, "r") as file:
        for line in file:
            ages.append(int(line.strip()))
    min_age = min(ages)
    max_age = max(ages)
    average_age = sum(ages) / len(ages)
    return min_age, max_age, average_age
filename = "ages.txt"
with open(filename, "w") as file:
    for _ in range(20):
        age = random.randint(18, 60)
        file.write(str(age) + "\n")
min_age, max_age, avg_age = survey_analysis(filename)
print("Eng kichik yosh:", min_age)
print("Eng katta yosh:", max_age)
print("O‘rtacha yosh:", round(avg_age, 2))
```

```python
rfilename = "feedback.txt"
with open(filename, "w") as file:
    file.write("The product is good and I like it.\n")
    file.write("Service was bad, but overall okay.\n")
    file.write("Food quality is average.\n")
    file.write("Good experience, will come again!\n")
    file.write("Not bad at all.\n")
def feedback_analyzer(filename):
    counts = {"good": 0, "bad": 0, "average": 0}
    with open(filename, "r") as file:
        for line in file:
            line_lower = line.lower()
            for word in counts.keys():
                counts[word] += line_lower.count(word)
    return counts
result = feedback_analyzer(filename)
print("Natija:")
for key, value in result.items():
    print(f"{key}: {value}")
```

```python

def unique_numbers(filename):
    file = open(filename, "r")
    sonlar = file.read().split()
    file.close()

    natija = []
    for son in sonlar:
        s = int(son)
        if s not in natija:
            natija.append(s)

    return natija


# test
f = open("raqamlar.txt", "w")
f.write("5\n3\n5\n7\n3\n9\n5")
f.close()

print(unique_numbers("raqamlar.txt"))
```

```python
def sales_filter(filename):
    file = open(filename, "r")
    savdolar = file.read().split()
    file.close()

    katta_savdolar = []
    for savdo in savdolar:
        s = int(savdo)
        if s > 1000:
            katta_savdolar.append(s)

    return katta_savdolar


# test
f = open("savdo.txt", "w")
f.write("500\n1500\n800\n2000\n1200\n300\n5000")
f.close()

print(sales_filter("savdo.txt"))
```

```python
def top_three(filename):
    file = open(filename, "r")
    sonlar = file.read().split()
    file.close()

    raqamlar = []
    for son in sonlar:
        raqamlar.append(int(son))

    raqamlar.sort()
    raqamlar.reverse()

    return [raqamlar[0], raqamlar[1], raqamlar[2]]


# test
f = open("sonlar.txt", "w")
f.write("45\n12\n89\n23\n67\n34\n90\n15")
f.close()

print(top_three("sonlar.txt"))
```
