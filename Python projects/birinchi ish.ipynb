```python
REPORT='report.txt'
SALES_FILE='sales.txt'
INVENTORY_FILE='inventory.txt'
EMPLOYEES_FILE='employees.txt'
hodimlar={'ali':{'maosh':5000000,'lavozim':'direktor'},
          'gayrat':{'maosh':3000000,'lavozim':'sotuvchi'}}
def mini_erp():
    try:
        with open(INVENTORY_FILE, 'r') as f:
            inventory=dict()
            for line in f:
                info=line.split(':')
                nom=info[0].strip().lower()
                miqdor=int(info[1].strip())
                narx=int(info[2].strip())
                inventory[nom]={'stock':miqdor,'price':narx}
    except FileNotFoundError:
        print('Kechirasiz file topilmadi')
    print('XUSH KEIBSIZ')
    print(4*'-','Mini erp tizimi','-'*4)
    print(inventory)


    while True:
        boshtanlov = None
        print('1.Mahsulotlarni boshqarish')
        print('2.Savdoni qayd etish')
        print('3.Hodimlarni boshqarish')
        print('4.Hisobotlar📦')
        print('5.Chiqish🔙')
        try:
            boshtanlov=input('1-5 gacha birini tanlang:').strip()
            if not boshtanlov:
                print('❌ hech nima kiritmadingiz,qayta urining')
                continue
            boshtanlov=int(boshtanlov.strip())
        except ValueError:
            print('❌ Faqat son kiriting')
        if boshtanlov == 1:

            def mahsulotlarni_boshqarish():
                choice=None
                while True:
                    print('1.Yangi mahsulot qo\'shish')
                    print('2.mahsulot zaxirasini yangilash')
                    print('3.5tadan kam maxsulotlarni ko\'rsatish')
                    print('4.chiqish')
                    try:
                        choice=int(input('1-4 gacha tanlovingizni kiriting: ').strip())
                        if not choice:
                            print('siz hech nima kiritmadingiz,qayta urining')
                    except ValueError:
                        print('❌Faqat son kiriting')

                    if choice == 1:

                        def mahsulot_qoshish():
                            print('chiqish uchun "stop"')
                            while True:
                                mahsulot=input('yangi mahsulot nomini kiriting: ').strip().lower()
                                if not mahsulot:
                                    print('🚫 siz hech qanday mahsulot nomini kiritmadingiz,qayta urining ')
                                elif mahsulot in inventory:
                                    print('🔴 omborda bu mahsulot mavjud')
                                elif mahsulot=='stop':
                                    return 'salomat bo\'ling'
                                else:
                                    if mahsulot not in inventory:
                                        break
                            print('chiqish uchun \'stop\'')
                            while True:
                                miqdor=input('sonini kiriting: ').strip().lower()
                                if not miqdor:
                                    print('🚫 siz hech qanday mahsulot miqdor kiritmadingiz,qayta urining')
                                elif miqdor=='stop':
                                    return 'salomat bo\'ling'
                                else:
                                    try:
                                        miqdor=int(miqdor)
                                        if miqdor<0:
                                            print('mahsulot soni butun son bolishi kerak')
                                        else:
                                            break
                                    except ValueError:
                                        print('❌ Faqat son kiriting')
                            print('chiqish uchun \'stop\'')
                            while True:
                                narx=input('mahsulot narxini kiriting: ').strip().lower()
                                if not narx:
                                    print('🚫 siz hech qanday mahsulot narxini kiritmadingiz,qayta urining ')
                                elif narx=='stop':
                                    return 'salomat bo\'ling'
                                else:
                                    try:
                                        narx=int(narx)
                                        if narx<=0:
                                            print('🚫 mahsulot narxi 0 dan kichik yoki teng bolishi kerakmas')
                                        else:
                                            break
                                    except ValueError:
                                        print('❌ Faqat son kiriting')
                            inventory[mahsulot]={'stock':miqdor,'price':narx}
                            with open(INVENTORY_FILE, 'w') as f:
                                for nom,info in inventory.items():
                                    f.write(f'{nom}:{info["stock"]}:{info["price"]}\n')
                            print('🎇 mahsulot muvaffaqiyatli qoshildi')
                            return f'nomi:{mahsulot},miqdor: {miqdor},narx: {narx}'
                        print(mahsulot_qoshish())
                    elif choice == 2:
                        def zaxira_yangilash():
                            while True:

                                print('chiqish uchun \'stop\'')
                                print('1.zaxiraga qo\'shish')
                                print('2.zaxiradan ayirish')
                                print('3.chiqish🔙')
                                try:
                                    ichkitanlov=int(input('3tasidan birini tanlang: ').strip())
                                    if not ichkitanlov:
                                        print('🚫 siz hech nima kiritmadingiz')
                                    else:
                                        if ichkitanlov==1:
                                            def qoshish():
                                                while True:
                                                    nom = input('mahsulot nomini kiriting(chiqish uchun "stop"): ').strip().lower()
                                                    if not nom:
                                                        print('🔴 siz qanday nom kiritmadingiz,qayta urining ')
                                                    elif nom == 'stop':
                                                        return 'salomat bo\'ling'
                                                    elif nom not in inventory:
                                                        print(f'{nom} <- bu nomli mahsulot omborda mavjud emas')
                                                    else:
                                                        break

                                                while True:
                                                        miqdor = input('miqdorni kiriting(chiqish uchun "stop"): ').strip()
                                                        if not miqdor:
                                                            print('🔴 Xato hech narsa kiritmadingiz')
                                                        elif miqdor=='stop':
                                                            return 'salomat bo\'ling'
                                                        else:
                                                            try:
                                                                miqdor=int(miqdor)
                                                                if miqdor <= 0:
                                                                    print('mahsulot miqdori 0 yoki undan kichkina bo\'lmasligi kerak')
                                                                else:
                                                                    break
                                                            except ValueError:
                                                                print('❌ faqat son kiriting')
                                                inventory[nom]['stock']+=miqdor
                                                with open(INVENTORY_FILE, 'w') as f:
                                                    for nom,info in inventory.items():
                                                        f.write(f'{nom}:{info["stock"]}:{info["price"]}\n')
                                                return f'🎯mahsulot miqdori muvafaqqiyatli qoshildi.mana sizga mahsulotlar ombori: {inventory}'
                                            print(qoshish())
                                        elif ichkitanlov==2:
                                            def ayirish():
                                                while True:
                                                    nom=input('mahsulot nomini kiriting(chiqish uchun "stop"): ').strip()
                                                    if not nom:
                                                        print('❌ Xato hech nima kiritmadingiz,qayta urining ')
                                                    if nom=='stop':
                                                        return 'salomat bo\'ling'
                                                    elif nom not in inventory:
                                                        print(f'{nom} <-bu nomli mahsulot omborda mavjud emas')
                                                    else:
                                                        break

                                                while True:
                                                    miqdor=input('miqdorni kiriting(chiqish uchun "stop"): ').strip().lower()
                                                    if not miqdor:
                                                        print('❌ Xato hech nima kiritmadingiz,qayta urining ')
                                                    elif miqdor=='stop':
                                                        return 'salomat bo\'ling'
                                                    else:
                                                        try:
                                                            miqdor=int(miqdor)
                                                            if miqdor <= 0:
                                                                print('miqdor 0 yoki undan kichkina bo\'lishi mumkinmas')
                                                            elif miqdor>inventory[nom]['stock']:
                                                                print(f'omborda bu mahsulot yetarli emas.info: {inventory[nom]['stock']} ta')
                                                            else:
                                                                if miqdor <= inventory[nom]['stock']:
                                                                    break
                                                        except ValueError:
                                                            print('❌ xato faqat son kiriting')

                                                inventory[nom]['stock']-=miqdor
                                                with open(INVENTORY_FILE, 'w') as f:
                                                    for nom,info in inventory.items():
                                                        f.write(f'{nom}:{info["stock"]}:{info["price"]}\n')
                                                return f'🎯mahsulot miqdori muvafaqqiyatli yangilandi: {inventory[nom]["stock"]} ta qoldi hozir'
                                            print(ayirish())
                                        elif ichkitanlov==3:
                                            return 'salomat bo\'ling'
                                        else:
                                            print('xato faqat menyudagini kiriting')
                                except ValueError:
                                    print('❌ Xato faqat son kiriting')
                                    continue
                        print(zaxira_yangilash())
                    elif choice == 3:
                        def mahsulot_korsatish():
                            lugat={}
                            for nom,info in inventory.items():
                                if info['stock']<5:
                                    lugat[nom]=info
                            if len(lugat)==0:
                                return 'tadan kam mahsulotlar yo\'q'
                            else:
                                return '5tadan kam mahsulotlar: {}'.format(lugat)
                        print(mahsulot_korsatish())
                    elif choice == 4:
                        return 'salomat bo\'ling'
                    else:
                        print('❌Xato! to\'g\'ri kiriting')
            print(mahsulotlarni_boshqarish())

        elif boshtanlov == 2:
            sales=list()
            while True:
                mahsulot=input('mahsulotni kiriting(chiqish uchun "stop"): ').strip().lower()
                if not mahsulot:
                    print('❌ hech nima kiritmadingiz,qayta urining ')
                elif mahsulot.isdigit():
                    print('faqat nom kiriting son emas')
                elif mahsulot=='stop':
                    return 'salomat bo\'ling'
                elif mahsulot not in inventory:
                    print('bu mahsulot omborda mavjud emas')
                else:
                    break
            while True:
                miqdor=input('miqdorni kiriting(chiqish uchun "stop"): ').strip().lower()
                if not miqdor:
                    print('❌ hech nima kiritmadingiz,qayta urining ')
                elif miqdor=='stop':
                    return 'salomat bo\'ling'
                else:
                    try:
                        miqdor=int(miqdor)
                        if miqdor>=inventory[mahsulot]['stock']:
                            print('omborda buncha miqdor mavjud emas,bori: {} ta qolgan'.format(inventory[mahsulot]['stock']))
                        elif miqdor <= 0:
                            print('❌miqdor 0 yoki undan kichik bolmasin')
                        else:
                            break
                    except ValueError:
                        print('❌ faqat son kiriting')
            inventory[mahsulot]['stock']-=miqdor
            sales.append([mahsulot,inventory[mahsulot]['stock'],inventory[mahsulot]['price']])
            with open(SALES_FILE, 'a') as f:
                for info in sales:
                    f.write(f'{info[0]}:{info[1]}:{info[2]}\n')
            print('✅muvafaqqiyatli yangilandi')
            print('siz qilgan savdo: {}'.format(sales))
            print(f'shu kungacha bolgan barcha savdolaringiz:')
            try:
                with open(SALES_FILE, 'r') as f:
                    print(f.read())
            except FileNotFoundError:
                print('🔴 kechirasiz tizimda muammo bor, keyinroq urining')
        elif boshtanlov == 3:
            def hodimlar():
                while True:
                    print('1.yangi hodim qo\'shish')
                    print('2.barcha maoshlarga foiz qo\'shish')
                    print('3.eng ko\'p maoshlilarni chiqarish')
                    print('4.o\'rtachadan yuqori maoshlilarni chiqarish')
                    print('5.chiqish')
                    try:
                        ichkichoice = int(input('tanlov: ').strip())
                        if not ichkichoice:
                            print('❌ hech nima kiritmadingiz,qayta urining ')
                        elif 0 >= ichkichoice > 5:
                            print('xato qayta kiriting')
                        else:
                            break
                    except ValueError:
                        print('❌ Xato faqat son kiriting ')

                if ichkichoice == 1:
                    def yangi_hodim_qoshish():
                        global hodimlar
                        while True:
                            ism = input('ismingiz: ').strip().lower()
                            if len(ism) == 0:
                                print('❌ Xato')
                            elif ism in hodimlar:
                                print('bu hodim bizda bor,boshqa ismingizni kiriting')
                            elif ism.isdigit():
                                print('❌ raqam kiritmang')
                            else:
                                break
                        while True:
                            try:
                                oylik=int(input('oylik: ').strip())
                                if oylik<100000:
                                    print('maosh 100mingdan kichkina bolmasin')
                                elif oylik>10000000:
                                    print('maosh 10mlndan katta bolmasin')
                                else:
                                    break
                            except ValueError:
                                print('❌Xato! faqat son kiriting')
                        while True:
                            kasb=input('lavozim: ').strip().lower()
                            if not kasb:
                                print('❌ xato!')
                            else:
                                break
                        hodimlar[ism]={'maosh':oylik,'lavozim':kasb}
                        return ism,'->',hodimlar[ism]

                    print(yangi_hodim_qoshish())
                elif ichkichoice == 2:

                    def barcha_maoshlarga_foiz_qosh():
                        global hodimlar
                        foiz=None
                        while True:
                            try:
                                foiz=int(input('foiz: ').strip())
                                if foiz<=0:
                                    print('0 dan katta son kiriting')
                                else:
                                    break
                            except ValueError:
                                print('❌ Xato faqat son kiriting')

                        foiz=foiz/100
                        print('oshishdan oldin: {}'.format(hodimlar))
                        vaqtincha={}
                        for ism,info in hodimlar.items():
                            vaqtincha[ism]={'maosh':info['maosh']*foiz,'lavozim':info['lavozim']}
                        hodimlar.clear()
                        import copy
                        hodimlar=copy.deepcopy(vaqtincha)
                        return hodimlar
                    print(barcha_maoshlarga_foiz_qosh())
                elif ichkichoice == 3:
                    def eng_kop_maosh_oladiganlar():
                        global hodimlar
                        engkop=max(hodimlar.items(),key=lambda x:x[1]['maosh'])
                        return 'eng ko\'p maosh oladigan hodim: {},uning maoshi: {} so\'m'.format(engkop[0],engkop[1]['maosh'])
                    print(eng_kop_maosh_oladiganlar())

                elif ichkichoice == 4:
                    def ortachadan_yuqori_maosh():
                        global hodimlar
                        jami=0
                        sanoq=0
                        for ism,info in hodimlar.items():
                            jami+=info['maosh']
                            sanoq+=1
                        ortacha=jami/sanoq
                        tahlil=[ism for ism,info in hodimlar.items() if info['maosh']>ortacha]
                        if not tahlil:
                            return 'o\'rtacha maoshdan yuqori oladigan hodim topilmadi'
                        return f'ortachadan yuqori maosh oladigan hodimlar: {tahlil}'
                    print(ortachadan_yuqori_maosh())
                else:
                    return 'salomat bo\'ling'
            print(hodimlar())

        elif boshtanlov == 4:
            tanlov=None
            while True:
                print('1.jami tushumni ko\'rish')
                print('2.eng kop sotilgan mahsulotni korish')
                print('3.eng kam mahsulot')
                print('4.tushumlarni ko\'rish')
                print('5.chiqish')
                try:
                    tanlov=int(input('tanlov: ').strip())
                    if not tanlov:
                        print('❌ Xato!')
                    else:
                        if tanlov==1 or tanlov==2 or tanlov==3 or tanlov==4 or tanlov==5 :
                            break
                        else:
                            continue
                except ValueError:
                    print('❌ Faqat son kiriting')
            if tanlov==1:
                def umumiy_tushum():
                    try:
                        with open(SALES_FILE,'r',encoding='utf-8') as f:
                            lst=list()
                            for line in f:
                                line=line.split(':')
                                nom=line[0].strip().lower()
                                son=int(line[1].strip())
                                narx=int(line[2].strip())
                                lst.append([nom,son,narx])
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    jami=0
                    for i in lst:
                        jami+=i[1]*i[2]
                    print( f'umumiy tushum: {jami} so\'m')
                    try:
                        with open(REPORT,'a',encoding='utf-8') as f:
                            f.write(f'Jami tushum: {jami}\n')
                    except FileNotFoundError:
                        print('fayl topilmadi')
                print(umumiy_tushum())
            elif tanlov==2:
                def eng_kop_sotilgan_mahsulot():
                    try:
                        with open(SALES_FILE,'r',encoding='utf-8') as f:
                            lugat=dict()
                            for line in f:
                                line=line.split(':')
                                nom=line[0].strip().lower()
                                son=int(line[1].strip())
                                narx=int(line[2].strip())
                                lugat[nom]=lugat.get(nom,0)+son
                            engkop=max(lugat.items(),key=lambda x:x[1])
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    try:
                        with open(REPORT,'a',encoding='utf-8') as f:
                            f.write(f'eng kop sotilgan mahsulot: {engkop}\n')
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    return f'eng ko\'p sotilgan mahsulot: {engkop[0]}, soni: {engkop[1]} ta'
                print(eng_kop_sotilgan_mahsulot())

            elif tanlov==3:
                def eng_kam_zaxira():
                    try:
                        with open(INVENTORY_FILE,'r',encoding='utf-8') as f:
                            eng_kam=dict()
                            for line in f:
                                line=line.strip().split(':')
                                nom=line[0].strip().lower()
                                son=int(line[1].strip())
                                narx=int(line[2].strip())
                                eng_kam[nom]={'miqdor':son,'narx':narx}
                            eng_kam=min(eng_kam.items(),key=lambda x:x[1]['miqdor'])
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    try:
                        with open(REPORT,'a',encoding='utf-8') as f:
                            f.write(f'zaxiradagi eng kam mahsulot: {eng_kam}\n')
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    return f'zaxiradagi eng kam mahsulot: {eng_kam}'
                print(eng_kam_zaxira())

            elif tanlov==4:
                def tushum_hisoboti():
                    try:
                        with open(INVENTORY_FILE,'r',encoding='utf-8') as f:
                            yangi=dict()
                            for line in f:
                                line=line.strip().split(':')
                                nom=line[0].strip().lower()
                                son=int(line[1].strip())
                                narx=int(line[2].strip())
                                tushum=son*narx
                                yangi[nom]=yangi.get(nom,0)+tushum
                            kamayish=sorted(yangi.items(),key=lambda x:x[1],reverse=True)
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    try:
                        with open(REPORT,'a',encoding='utf-8') as f:
                            f.write(f'{kamayish}\n')
                    except FileNotFoundError:
                        print('fayl topilmadi')
                    return f'barcha tushumlar: {kamayish}'
                print(tushum_hisoboti())
            else:
                continue
        else:
            print('❌ xato! menyudagilarni kiriting')

print(mini_erp())
```
