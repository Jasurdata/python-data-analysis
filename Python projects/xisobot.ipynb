```python
def hisobot_chiqar():

    xarajat=dict()

    print('xarajat hisoblagichga xush kelibsiz')
    print('xarajatlaringizni kiriting.chiqish uchun \'tugadi\' deb yozing')
    while True:
        kiritish=input('xarajat(format: xarajat nomi summa): ')
        if kiritish.lower()=='tugadi':
            try:
                with open('xarajat.txt','r') as f:
                    lugat=dict()
                    for i in f:
                        i=i.split(':')
                        nom=i[0].strip()
                        summa=int(i[1])
                        lugat[nom]=lugat.get(nom,0)+summa
            except FileNotFoundError:
                print('bunday fayl topilmadi')
                yangi=dict()
                for key in xarajat|lugat:
                    yangi[key]=xarajat.get(key,0)+lugat.get(key,0)
                with open('xarajat.txt','w') as f:
                    for key,value in yangi.items():
                        f.write(f'{key}:{value}\n')
            return f'bugun kiritgan xarajatlaringiz: {xarajat}\njami xarjatlaringiz: {lugat}'
        bolish = kiritish.split()
        if not bolish:
            print('xato! hech qanday xarjat kiritmadingiz,qayta urining')
        elif len(bolish)==1 or len(bolish)>2:
            print('xato! togri formatda kitmadiz.(format: xarajat_nomi summa)')
        else:
            nom=bolish[0].strip().lower()
            summa=bolish[1].strip()
            try:
                summa=int(summa)
                if summa<=0:
                    print('xato! summa 0dan kichik summa kiritmang')
            except ValueError:
                print('summani togri formatda kiriting')
            else:
                xarajat[nom]=xarajat.get(nom,0)+summa
                with open('xarajat.txt','a') as f:
                    f.write(f'{nom}:{summa}\n')
print(hisobot_chiqar())
```
