```python
def belgi_sanash(matn='bosh',harf='bosh'):
    if matn=='bosh' or harf=='bosh':
        return 'matn yoki harf yoki 2alasiham kiritilmagan,toliq kiriting'
    if not isinstance(matn,str) or not isinstance(harf,str):
        return 'Ikkala argument ham matn (str) bo‘lishi kerak'
    matn=matn.lower()
    harf=harf.lower()
    harf=str(harf)
    son=0
    if len(harf)>1:
        return 'faqat 1ta belgi kiriting'
    for i in matn:
        if i==harf:
            son+=1
    return son
print(belgi_sanash())
```

```python
def eng_uzun_ism(royxat='bosh'):
    if royxat=='bosh':
        return 'royxat kiritilmagan'
    cleaned=list(filter(lambda x: not isinstance(x,(int,float,bool)) and x.isalpha(),royxat))
    if not cleaned:
        return 'talabga javob beradigan malumot yoq'
    tartib=sorted(cleaned,key=lambda x: len(x),reverse=True)[0]
    return f'eng uzun ism bu-{tartib}'
print(eng_uzun_ism(['Adham','zahiriddin']))


def eng_uzun_ism(royxat='bosh'):
    if royxat=='bosh':
        return 'royxat kiritilmagan'
    uzun=''
    for i in royxat:
        if not isinstance(i,str) or not i.isalpha():
            continue
        if len(i)>len(uzun):
            uzun=i
    return f'eng uzun ism bu-{uzun}'
print(eng_uzun_ism(['Adham','44444444444','zahiriddin']))
```

```python
def yigindi():
    qoshuv=0
    while True:
        son=input('son kiriting(toxtash uchun "done"): ')
        if son=='done':
            break
        try:
            son=float(son)
        except:
            print('faqat son kiriting yoki done deb yozing')
            continue
        qoshuv+=son
    return f'yigindi: {qoshuv}'
print(yigindi())
```

```python
def juf(*son):
    if len(son)!=10:
        return 'faqat aniq 10ta butun son kiriting'
    juft=0
    toq=0
    for i in son:
        if isinstance(i, str) and i.isdigit():
            i = int(i)
        if not isinstance(i,int):
            return 'faqat butun son kiriting'
        if i%2==0:
            juft+=1
        if i%2==1:
            toq+=1
    return f'juft sonlar: {juft}ta, toq sonlar:{toq}ta'
print(juf(1,2,3,4,9,99,'90',4,5,5))
```

```python
def yulduzli_minora(*son):
    if len(son)!=1:
        return 'xato! faqat 1ta butun son kiriting'
    son=son[0]
    if isinstance(son,str):
        son=int(son)
    if not isinstance(son,int):
        return 'faqat butun son kiriting'
    for i in range(1,son+1):
        print(i*'*')
    return ''

print(yulduzli_minora(6))
```

```python
def raqamli_minora(*son):
    if len(son)!=1:
        return 'faqat 1ta butun son kiriting'
    son=son[0]
    try:
        if isinstance(son,bool):
            return 'faqat 1ta butun son kiriting'
        son=int(son)
        if son<0:
            return 'xato! faqat 0dan katta butun son kiriting'
        for i in range(1, son + 1):
            for j in range(1, i + 1):
                print(j, end='')
            print()
    except(ValueError,TypeError):
        return 'faqat butun son kiriting'
    return 'minora Tayyor'
print(raqamli_minora(8))
```

```python
def eng_kichik():
    min=list()
    while True:
        son=input('son kiriting(toxtash uchun "stop"): ')
        if son.lower()=='stop':
            break
        try:
            son=int(son)
            min.append(son)
        except:
            print('xato faqat butun son kiriting! ')
    cleaned=sorted(min)
    if not cleaned:
        return 'xato!umuman butun son kiritmadingiz'
    return f'eng kichik son->{cleaned[0]}'
print(eng_kichik())
```

```python
def kopaytirish_jadvali():
    natija=''
    for i in range(1,11):
        for j in range(1,11):
            natija+=f'{j}*{i}={i*j}'
        natija+='\t'
    return natija
print(kopaytirish_jadvali())
```

```python
def piramida(n):
    natija = ''
    for i in range(1, n+1):
        # chap tomondagi bo'sh joylar
        bosh = " " * (n - i)
        # yulduzchalar
        yulduz = "*" * (2*i - 1)
        # natijani chiqarish
        natija+=bosh + yulduz
        natija+='\n'
    return natija
print(piramida(10))
```
