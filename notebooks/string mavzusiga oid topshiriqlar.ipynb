```python
def unli_sanash(matn=None):
    if matn is None:
        return 'Xato! hech qanday matn kiritmagansiz'
    matn=str(matn)
    unli='aoiue'
    son=0
    for i in range(len(matn)):
        if matn[i] in unli:
            son+=1
        continue
    return f'kiritgan matningizda {son} ta unli harf bor'
print(unli_sanash())
```

```python
def eng_uzun_soz(soz=None):
    if soz is None:
        return 'Xato! siz hech qanday matn kiritmadingiz'
    soz=str(soz)
    bolish=soz.split(' ')
    eng_uzun=sorted(bolish, key=len, reverse=True)[0]
    uzunlik=len(eng_uzun)
    return f'kiritgan matningizdagi eng uzun so\'z: {eng_uzun}, uzunligi: {uzunlik} '
print(eng_uzun_soz('salom Python Mustaqillik'))
```

```python
def raqamni_shifrla(raqam=None):      #1-usul
    if raqam is None:
        return 'Xato! raqam kiritmadingiz'
    raqam=str(raqam)
    sanoq=0
    natija=''
    for i in raqam:
        sanoq+=1
        if sanoq<=4:
            natija+=i
        if sanoq>4:
            natija+='*'
    return natija
print(raqamni_shifrla('998993155581'))

def raqamni_shifrla(raqam=None):       #2-usul
    if raqam is None:
        return 'Xato! raqam kiritmadingiz'
    raqam=str(raqam)
    ish=(len(raqam)-4)*'*'+raqam[-5:-1]
    return ish
print(raqamni_shifrla('99890455556'))
```

```python
def substring(matn=None,qidiruv=None):
    if matn is None or qidiruv is None:
        return 'kerakli paramterlarni kiritmadingiz'
    matn=matn.lower()
    qidiruv=qidiruv.lower()
    bolish=matn.split(' ')
    tekshirish=list(filter(lambda x:x==qidiruv, bolish))
    if len(tekshirish)>0:
        return True
    return False
print(substring("Mening vatanim ozbekiston","mening"))
```

```python
def passport_qismlari(passport:str):   #1-usul
    passport = str(passport)
    if len(passport)!=9:
        return 'noto\'g\'ri kiritdingiz tekshirib qaytadan kiriting'
    natija=passport[:2]+len(passport[3:7])*'*'+passport[6:]
    return natija
print(passport_qismlari('AD2888115'))


def passport_qismlari(passport:str):   #2-usul
    passport = str(passport)
    if len(passport)!=9:
        return 'Xato! passport id 9ta belgi bolishi kerak'
    if not passport.startswith('AD' or 'AB' or 'AA'):
        return 'bosh harflarni noto\'g\'ri kiritdingiz'
    if not passport[2:].isdigit():
        return """Xato! boshidagi 2ta belgidan keyin faqat raqam bo\'lishi kerak"""
    sanoq=0
    natija=''
    for i in passport:
        if sanoq<2:
            natija+=i
        if 2<=sanoq<=4:
            natija+='*'
        if sanoq>5:
            natija+=i
        sanoq += 1
    return natija
print(passport_qismlari('AC28885'))
```

```python
def email_domen(email=None):
    if email is None:
        return 'Xato! siz hech qanday email kiritmadingiz'
    if ('1' or '2' or '3' or '4' or '5' or '6' or '7' or '8' or '9' or '0' ) not in email  or '@gmail.com' not in email or not len(email)>13:
        return 'email manzilingizni togri kiritmadingiz'
    email=str(email)
    qidirish=email.split('@')
    return qidirish[1]
print(email_domen('rjasur421@gmail.com'))
```

```python
def hashtag(*jumla):
    if len(jumla) == 0:
        return 'xato! hech qanday matn kiritmadingiz'
    natija=' '
    for i in jumla:
        i=str(i)
        i=i.split()
        for b in i:
            b=b[0]
            natija+=b
    natija='#'+natija.upper()
    return natija
print(hashtag('men pythonni sevaman','salom'))
```
