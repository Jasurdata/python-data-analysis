```python
import sqlite3
from datetime import datetime
conn=sqlite3.connect('sales.db')
cur=conn.cursor()
cur.execute("""CREATE TABLE IF NOT EXISTS Transactions(
        id integer primary key autoincrement,
        customer text,
        email text,
        product text,
        quantity integer,
        price float,
        total float
        )""")


def add_sale():


    while True:
        mijoz_ismi=None
        email=None
        mahsulot_nomi=None
        miqdor=None
        narx=None
        print('chiqish uchun \'0\' bosing')
        while True:
            mijoz_ismi=input("Enter mijoz ismi: ").strip().lower()
            if not mijoz_ismi:
                continue
            elif mijoz_ismi=='0':
                break
            elif mijoz_ismi.isdigit():
                print('Xato! ismda son qatnashmasin')
                continue
            else:
                break
        if mijoz_ismi=='0':
            break
        while True:
            email=input('email: ').strip()
            if not email:
                continue
            elif email=='0':
                break
            elif not email.endswith('@gmail.com') or '@' not in email or len(email)<8:
                print('emailda @gmail.com qatnashishi va @ bu belgi bolishi va uzunligi 8dan balanda bolishi kerak,qayta urining')
                continue
            else:
                break
        if email=='0':
            break
        while True:
            mahsulot_nomi=input('mahsulot_nomi: ').strip().title()
            if not mahsulot_nomi:
                continue
            elif mahsulot_nomi=='0':
                break
            else:
                break
        if mahsulot_nomi=='0':
            break
        while True:
            miqdor=input('miqdor: ').strip()
            if not miqdor:
                continue
            elif miqdor=='0':
                break
            try:
                miqdor=int(miqdor)
                if miqdor<0:
                    print('miqdor musbat son bolishi kerak')
                    continue
                else:
                    break
            except ValueError:
                print('Xato! faqat son kiriting')
                continue
        if miqdor=='0':
            break
        while True:
            narx=input('narx: ').strip()
            if not narx:
                continue
            elif narx=='0':
                break
            try:
                narx=float(narx)
                if narx<0:
                    print('narx musbat son bolishi kerak')
                    continue
                else:
                    break
            except ValueError:
                print('Xato! faqat son kiriting')
                continue
        if narx=='0':
            break
        cur.execute("""insert into Transactions(customer,email,product,quantity,price,total) values (?,?,?,?,?,?)""",
                    (mijoz_ismi, email, mahsulot_nomi, miqdor, narx,miqdor*narx))
        conn.commit()
        jami=miqdor*narx
        print(f'{mijoz_ismi} uchun savdo muvafaqqiyatli qo\'shildi.Jami: ${jami}')
        tushum += jami
        son+=1
        ism[mijoz_ismi]=ism.get(mijoz_ismi,0)+jami

def view_summer_y():
    cur.execute('select * from Transactions')
    data=len(cur.fetchall())
    cur.execute('select sum(total),avg(price) from Transactions')
    info=cur.fetchone()
    cur.execute('select customer,total from Transactions order by total desc limit 1')
    infor=cur.fetchone()
    cur.execute('select product,count(product) as soni from Transactions group by product order by soni desc limit 1')
    respond=cur.fetchone()
    print(f'Umumiy savdolar soni: {data}')
    print(f'Umumiy tushum: {info[0]}')
    print(f'o\'rtacha buyurtma: {info[1]}')
    print(f'Eng faol mijoz: {infor[0]}, {infor[1]}')
    print(f'eng mashhur mahsulot: {respond[0]}')


def search_customer():
    ism=None
    while True:
        ism=input('qidirilaotgan mijoz ismini kriting: ').strip().lower()
        if not ism:
            continue
        elif ism=='0':
            break
        elif ism.isdigit():
            print('xato! ismda raqam qatnashmasin')
            continue
        else:
            cur.execute('select customer,product,count(product) as soni,sum(price) from Transactions where customer=? group by customer,product',(ism,))
            data=cur.fetchall()
            if not data:
                print('bu mijoz bo\'yicha ma\'lumot topilmadi')
            else:
                if len(data)>1:
                    print(f'{ism} natijalari: ')
                    for i in data:
                        print(f'{i[1]} {i[2]} dona ${i[3]}')
                    break
                elif len(data)==1:
                    print(f'{ism} natijalari: `')
                    print(f'{data[1]} {data[2]} dona ${data[3]}')
                    break


def save_report():
    global son
    global tushum
    global ism

    sana=datetime.today()
    faol=max(ism.items(),key=lambda x: x[1])
    with open('report.txt','a',encoding='utf-8') as f:
        f.write('sales Report data Nova\n')
        f.write(f'{sana}\n')
        f.write(f'Umumiy savdolar: {str(son)}\n')
        f.write(f'umumiy tushum: ${str(tushum)}\n')
        f.write(f'eng faol mijoz: {str(faol[0])} {str(faol[1])}\n')
    print('Hisobot saqlandi.Xayr')



while True:
    print('\n===Data Nova Sales Tracker===')
    print('1.yangi savdo qo\'shish')
    print('2.statistika ko\'rish')
    print('3.mijozni qidirish')
    print('4.chiqish')
    tanlov=input('tanlang: ').strip()
    if not tanlov:
        continue
    try:
        tanlov=int(tanlov)
    except ValueError:
        print('Xato! faqat son kiriting!')
        continue
    if tanlov==1:
        add_sale()
    elif tanlov==2:
        view_summer_y()
    elif tanlov==3:
        search_customer()
    elif tanlov==4:
        save_report()
        break
    else:
        print('noto\'g\'ri tanlov qayta urinib ko\'ring')


cur.close()
conn.close()
```
