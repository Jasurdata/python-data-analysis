```python
INVENTORY_FILE = "inventory.txt"
EMPLOYEES_FILE = "employees.txt"
SALES_FILE = "sales.txt"
REPORT_FILE = "report.txt"
inventory = {}
sales = []
total_revenue = {}
employees = {}
def load_inventory():
    try:
        with open(INVENTORY_FILE, "r") as f:
            for line in f:
                name, stock, price = line.strip().split(",")
                inventory[name] = {"stock": int(stock), "price": float(price)}
    except FileNotFoundError:
        inventory.update({ "Laptop": {"stock": 10, "price": 800}, "Phone": {"stock": 20, "price": 600} })
def save_inventory():
    with open(INVENTORY_FILE, "w") as f:
        for name, info in inventory.items():
            f.write(f"{name},{info['stock']},{info['price']}\n")
def load_employees():
    try:
        with open(EMPLOYEES_FILE, "r") as f:
            for line in f:
                name, salary = line.strip().split(",")
                employees[name] = int(salary)
    except FileNotFoundError:
        employees.update({"Ali": 1200, "Madina": 1500})
def save_employees():
    with open(EMPLOYEES_FILE, "w") as f:
        for name, salary in employees.items():
            f.write(f"{name},{salary}\n")
def load_sales():
    try:
        with open(SALES_FILE, "r") as f:
            for line in f:
                name, qty, total = line.strip().split(",")
                sales.append((name, int(qty), float(total)))
                total_revenue[name] = total_revenue.get(name,0)+float(total)
    except FileNotFoundError:
        pass
def save_sales():
    with open(SALES_FILE, "w") as f:
        for name, qty, total in sales:
            f.write(f"{name},{qty},{total}\n")
def manage_products():
    while True:
        print("\n--- MAHSULOTLAR ---")
        print("1. Yangi mahsulot qo‘shish")
        print("2. Miqdorni yangilash")
        print("3. Kam zaxira (<=5)")
        print("4. Orqaga")
        try:
            choice = input("Tanlov: ")
        except KeyboardInterrupt:
            print("\nAsosiy menyuga qaytildi")
            break
        if choice == "1":
            try:
                nom = input("Nomi: ").title()
                miqdor = int(input("Miqdori: "))
                narx = float(input("Narxi: "))
                inventory[nom] = {"stock": miqdor, "price": narx}
                save_inventory()
                print("Mahsulot qo‘shildi")
            except:
                print("Xato kiritildi")
        elif choice == "2":
            try:
                nom = input("Nomi: ").title()
                miqdor = int(input("Qo‘shish/ayirish: "))
                if nom in inventory:
                    inventory[nom]["stock"] += miqdor
                    save_inventory()
                    print("Miqdor yangilandi")
                else:
                    print("Mahsulot topilmadi")
            except:
                print("Xato kiritildi")
        elif choice == "3":
            for n, info in inventory.items():
                if info["stock"] <= 5:
                    print(f"{n} - {info['stock']} dona")
        elif choice == "4":
            break
        else:
            print("Xato tanlov")
def manage_sales():
    try:
        nom = input("Mahsulot nomi: ").title()
        qty = int(input("Miqdor: "))
        if nom not in inventory:
            print("Mahsulot yo‘q")
            return
        if inventory[nom]["stock"] < qty:
            print("Yetarli mahsulot yo‘q")
            return
        jami = qty * inventory[nom]["price"]
        inventory[nom]["stock"] -= qty
        sales.append((nom, qty, jami))
        total_revenue[nom] = total_revenue.get(nom, 0) + jami
        save_inventory()
        save_sales()
        print(f"Savdo: {qty} x {nom} = {jami}$")
    except:
        print("Xato kiritildi")
def manage_employees():
    while True:
        print("-- HODIMLAR ---")
        print("1. Hodim qo‘shish")
        print("2. Maoshlarni oshirish (%)")
        print("3. Eng ko‘p maosh oluvchi")
        print("4. O‘rtachadan yuqori maosh oluvchilar")
        print("5. Orqaga")
        try:
            t = input("Tanlov: ")
        except KeyboardInterrupt:
            print("\nAsosiy menyuga qaytildi")
            break
        if t == "1":
            try:
                ism = input("Ism: ").title()
                salary = int(input("Maosh: "))
                employees[ism] = salary
                save_employees()
                print("Hodim qo‘shildi")
            except:
                print("Xato kiritildi")
        elif t == "2":
            try:
                foiz = float(input("Foiz: "))
                for e in employees:
                    employees[e] += employees[e] * foiz / 100
                save_employees()
                print("Maoshlar oshirildi")
            except:
                print("Xato kiritildi")
        elif t == "3":
            if employees:
                eng_kop = max(employees, key=employees.get)
                print(f"Eng ko‘p maosh: {eng_kop} - {employees[eng_kop]}$")
            else:
                print("Hodimlar yo‘q")
        elif t == "4":
            if employees:
                avg = sum(employees.values())/len(employees)
                yuqori = [e for e, s in employees.items() if s>avg]
                print("O‘rtachadan yuqori:", yuqori)
            else:
                print("Hodimlar yo‘q")
        elif t == "5":
            break
        else:
            print("Xato tanlov")
def show_reports():
    print("-- HISOBOT ---")
    jami = sum(total_revenue.values())
    print("Umumiy tushum:", jami)
    if total_revenue:
        # DSU: eng ko‘p sotilgan mahsulot
        sorted_sales = sorted(total_revenue.items(), key=lambda x:x[1], reverse=True)
        print("Eng ko‘p sotilgan:", sorted_sales[0][0], "-", sorted_sales[0][1])
    if inventory:
        kam = min(inventory.items(), key=lambda x:x[1]['stock'])
        print("Eng kam zaxira:", kam[0], "-", kam[1]['stock'])

    if total_revenue:
        print("Savdolar (tushum bo‘yicha kamayish):")
        for p, t in sorted_sales:
            print(f"{p} - {t}$")
    with open(REPORT_FILE, "w") as f:
        f.write(f"Umumiy tushum: {jami}$\n")
        if total_revenue:
            f.write(f"Eng ko‘p sotilgan: {sorted_sales[0][0]} - {sorted_sales[0][1]}$\n")
        if inventory:
            f.write(f"Eng kam zaxira: {kam[0]} - {kam[1]['stock']}\n")
        if total_revenue:
            f.write("Savdolar (tushum bo‘yicha kamayish):\n")
            for p, t in sorted_sales:
                f.write(f"{p} - {t}$\n")
    print("Hisobot report.txt fayliga saqlandi")
load_inventory()
load_employees()
load_sales()
while True:
    print("\n=== MINI ERP ===")
    print("1. Mahsulotlarni boshqarish")
    print("2. Savdoni qayd etish")
    print("3. Hodimlarni boshqarish")
    print("4. Hisobotlar")
    print("5. Chiqish")
    try:
        choice = input("Tanlovingiz: ")
    except KeyboardInterrupt:
        print("\nDastur tugadi")
        break
    if choice == "1":
        manage_products()
    elif choice == "2":
        manage_sales()
    elif choice == "3":
        manage_employees()
    elif choice == "4":
        show_reports()
    elif choice == "5":
        print("Dastur tugadi")
        break
    else:
        print("Xato tanlov")
```
